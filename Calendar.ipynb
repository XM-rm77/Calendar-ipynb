```python
def Ly (year):
  cy = 'common year'
  ly = 'leap year'
  if (year % 4 != 0):
    return cy
  elif (year % 100 != 0):
    return ly
  elif (year % 400 != 0):
    return cy
  elif (year % 400 == 0 and year < 3200):
    return ly 
  elif (year % 3200 != 0):
    return ly
  else:
    return cy
print('enter the year')
print(Ly(int(input())))
```

```python
def DoW (day, month, year):
  if (month == 1): #If January, change month to 13 and subtract 1 from year
    month = 13
    year = year - 1
  if (month == 2) : #If February, change month to 14 and subtract 1 from year
    month = 14
    year = year - 1
  K = year % 100
  J = K // 100
  h = ((day + (13 * (month + 1)) // 5) + K + (K // 4) + (J // 4) - 2*J) % 7
  return h
print('Enter the day')
d = int(input())
print('Enter the month')
m = int(input())
print('Enter the year')
y = int(input())
h = DoW(d,m,y)
iso = ((h + 5) % 7) + 1
if (iso == 1):
  print('Monday')
elif (iso == 2):
  print('Tuesday')
elif (iso == 3):
  print('Wednesday')
elif (iso == 4):
  print('Thursday')
elif (iso == 5):
  print('Friday')
elif (iso == 6):
  print('Saturday')
else:
  print('Sunday')
```
